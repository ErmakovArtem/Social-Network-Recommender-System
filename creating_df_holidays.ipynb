```python
import holidays
import pandas as pd

from database import postgres_connection
```

```python
rows = []

def add(country, dates):
    for d, name in dates:
        rows.append({
            "country": country,
            "date": d,
            "holiday_name": name
        })

year_list = [2021, 2022, 2023, 2024, 2025, 2026]
country_list = ['Russia', 'Ukraine', 'Belarus', 'Kazakhstan', 'Turkey', 'Finland', 'Azerbaijan', 'Switzerland', 'Estonia', 'Cyprus']

for country in country_list[:]:
    for year in year_list[:]:
        for date in holidays.country_holidays(country, years=year).keys():
            rows.append({
                'country': country,
                'date': date,
            })

df_holiday = pd.DataFrame(rows)
df_holiday["date"] = pd.to_datetime(df_holiday["date"])
```
